```python
############### Import modules and basic functions ###########
import pandas as pd
from math import radians, cos, sin, asin, sqrt
import random as rd
import ast
import numpy as np
from sys import stdout
import time
from scipy.spatial import distance_matrix
import matplotlib.pyplot as plt
import matplotlib
from collections.abc import Iterable
import os
import glob
import scipy.interpolate as interpolate

current_dir=os.getcwd()
def flatten(lis):
    """
    convert complex nested list to single list
    """
    for item in lis:
        if isinstance(item, Iterable) and not isinstance(item, str):
            for x in flatten(item):
                yield x
        else:        
            yield item
def distance(lon1, lat1, lon2, lat2):
    """
    Calculate the distance between two coordinate
    Great circle of the earth should be considered later in wide range of study
    """

    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    return np.sqrt(dlon**2+dlat**2)
```

# Sampling reservoir faults along the ray path in Groningen
### Since fast-axis direction ($\varphi$) in shear-wave splitting correlates with maximum horizontal stress($S_{Hmax}$) and Groningen is a well-known field that has a normal fault regime, we can consider the fault strike directions as $S_{Hmax}$ and compare with measured $\varphi$. 
### Subsequent figure and code samples demonstrate how we sample the fault along the ray path we used.
![nn](./Fault_sampling_figure.png)


```python
def fault_sampling(summ_dir=current_dir+'/Groningen_1420_MAN.summ', F_dir=current_dir+'/FaultModel_geometries.dat'):
    """
    Collect the fault orientations along the ray path.
    Regarding most faults are located at the reservoir depth, used a raypath surrounding ellipse as a range of fault sampling. 
    Because faults mostly located at the reservoir and all events occured at the reservoir depth, ellipse focuses were determined as event epicenter and midpoint of event-station. 
    """
    
    
    #Read the raypath data from the shear wave splitting measurement (event-station pairs)
    summ=pd.read_csv(summ_dir, header=None)
    
    #Read the fault data in Groningen
    F=pd.read_csv(F_dir)
    
    F['StrikeAzimuth']=""    #strike direction

    F['StrikeAzimuth']=(F['DipAzimuth']+90)%360
    F['dipcc']=(450-F['DipAzimuth'])%360
    summ.rename(columns={list(summ)[0]:'evid', 
                         list(summ)[1]:'station',
                         list(summ)[2]:'stlat',
                         list(summ)[3]:'stlon', 
                         list(summ)[5]:'year', 
                         list(summ)[7]:'latitude',
                         list(summ)[8]:'longitude',
                         list(summ)[20]:'dt',
                         list(summ)[22]:'phi'}, inplace=True)
    fault=F[['lon', 'lat', 'StrikeAzimuth']]
    summ=summ[['stlon', 'stlat', 'latitude', 'longitude', 'phi']]
    summ['fault']=""

    #########Save all fault directions along the raypath (with ellipse sampling)###########
    start=time.time()
    for i in range(len(summ)):

        lat=summ.loc[i, 'latitude']   #event location (Focus #1 of the ellipse)
        lon=summ.loc[i, 'longitude']
        stlat=summ.loc[i, 'stlat']    #station location
        stlon=summ.loc[i, 'stlon']
        mdlat=(lat+stlat)/2           #event-station midpoint location  (Focus #2 of the ellipse)
        mdlon=(lon+stlon)/2
        fau=[]                        
        d=distance(lon, lat, mdlon, mdlat)        #distance d=|F1-F2|=2ae => 2a=d/e (if a==semi-major axis, e==eccentricity)
        e=0.97                  #select the eccentricity of the ellipse

        ####### Find the fault point located in innerside of the ellipse######
        # Because the fault data is very large (>35000), we use matrix to calculate the distances(d1=|F1-faultpoint|, d2=|F2-faultpoint|)
        d12=distance_matrix(fault[['lon', 'lat']], [[lon, lat], [mdlon, mdlat]])
        
        # Simple mathematical trick to easily find data point inside of the ellipse
        ids=np.squeeze(np.argwhere((d12[:, 0]+d12[:,1])<d/e))   #d1+d2<2a,  2ae=d => d1+d2<d/e
        #save list of sampled faults' orientation in each raypath
        summ['fault'].iloc[i]=fault.iloc[ids].StrikeAzimuth.tolist()
    return summ



def draw_histogram(summfile):
    fault_data=np.array(list(flatten(summ['fault'].tolist())))
    
    #### convert the fault direction coordinate to (-90 ~ 90) ######
    for i in range(fault_data.size):
        strike=fault_data[i]
        if (strike>90)&(strike<=270):
            ori=strike-180
        elif (strike>270)&(strike<=360):
            ori=strike-360
        else:
            ori=strike
        fault_data[i]=ori
    ############ Plot Figures ################
    phi=summ.phi
    fig, ax = plt.subplots(2, 1,figsize=(8,10))
    ax[0].hist(fault_data, label='Sampled fault strike', color='coral', linewidth=2.0, edgecolor='black', bins=18)
    ax[1].hist(summ.phi, label=r'Shear-wave splitting $\varphi$', color='dimgray', linewidth=2.0, edgecolor='black', bins=18)
    ax[1].set_ylim([0, 3500])
    ax[1].set_ylim([0, 60])
    ax[1].set_xlabel(r'Direction ($\degree$)', fontsize=15)
    for i in range(len(ax)):
        ax[i].set_xticks(range(-90,100,20))
        ax[i].legend(fontsize=15, loc='upper right')
        ax[i].tick_params(axis='both', which='major', labelsize=15)
        ax[i].set_xlim([-100,100])
        ax[i].set_ylabel('Number of measurements', fontsize=15)
```

```python
summ=fault_sampling()
draw_histogram(summfile=summ)
```

### If we spatially average those data, we can compare spatial distribution of sampled fault with $\varphi$.  
### I revised spatial averaging code from TESSA (Johnson et al., 2011) with MATLAB. However, due to this MATLAB code is too long to be shared, here we will skip the process in MATLAB and take its results. 
### Instead, I provide simple code that makes comparison between spatial distribution of fault data and $\varphi$.

```python
def cosine_diff(loc_phi=current_dir+'/SA', loc_fault=current_dir+'/syn', loc_field=current_dir+'/Gasfield2.txt'):
    path_phi=glob.glob(loc_phi+'/*.loc')       #Averaged Phi in each grid cell
    path_fault=glob.glob(loc_fault+'/*.loc')   #Averaged fault strike in each grid cell
    phi=pd.DataFrame(columns=['lon', 'lat', 'mean','err', 'dev'], index=range(len(path_phi)))
    fault=pd.DataFrame(columns=['lon', 'lat', 'mean','err', 'dev'],index=range(len(path_fault)))
    
    #MAKE DATAFRAMES to COMPARE phi and fault strike#
    for i in range(len(path_phi)):
        num=path_phi[i].split('/')[-1].split('.')[0]
        num=''.join(c for c in num if c.isdigit())
        loc_file=pd.read_csv(path_phi[i], sep=' ',header=None, names=['lon', 'lat'], skipinitialspace=True).loc[0]
        phi_file=pd.read_csv("/".join(path_phi[i].replace('\\','/').split('/')[:-1])+'/phi'+num+'.tmp',sep=' ',header=None, names=['mean', 'err', 'dev'], skipinitialspace=True).loc[0]
        lon=loc_file[0]; lat=loc_file[1]; mean=phi_file[0]; err=phi_file[1]; dev=phi_file[2]
        phi.loc[i]=[lon, lat, mean, err, dev]

    for i in range(len(path_fault)):
        num=path_fault[i].split('/')[-1].split('.')[0]
        num=''.join(c for c in num if c.isdigit())

        loc_file=pd.read_csv(path_fault[i], sep=' ',header=None, names=['lon', 'lat'], skipinitialspace=True).loc[0]
        phi_file=pd.read_csv("/".join(path_fault[i].replace('\\', '/').split('/')[:-1])+'/phi'+num+'.tmp',sep=' ',header=None, names=['mean', 'err', 'dev'], skipinitialspace=True).loc[0]
        lon=loc_file[0]; lat=loc_file[1]; mean=phi_file[0]; err=phi_file[1]; dev=phi_file[2]
        fault.loc[i]=[lon, lat, mean, err, dev]

    #########################################################
    ################ Calculate cosine difference ############

    compare=pd.merge(phi, fault, on=['lon', 'lat'])
    compare['diff']=""
    for i in range(len(compare)):
        c=compare.loc[i]
        if (c.err_x<25)&(c.err_y<25)&(c.dev_x<25)&(c.dev_y<25): #calculate cosine difference if standard deviation and error is low
            compare['diff'].loc[i]=abs(np.cos(np.deg2rad(c.mean_x-c.mean_y)))   #0 (orthogonal) < cos|phi-strike| <1 (parallel)
        else: compare['diff'].loc[i]=100                      
    compare=compare[compare['diff']<100]    #Exclude if the grid cell has too deviated data distribution  
    compare[['lon', 'lat', 'diff']].to_csv(loc_fault+'/Table_diff.dat', header=None, sep=' ', index=None)  #Save difference for GMT figure
    Gasfield=pd.read_csv(loc_field, sep='\t', names=['x', 'y'])
    xmin = compare['lon'].min()
    xmax= compare['lon'].max()
    ymin = compare['lat'].min()
    ymax = compare['lat'].min()
    x= compare['lon']
    y= compare['lat']
    z= compare['diff']
    xx = np.linspace(x.min(), x.max(), len(set(x)))
    yy = np.linspace(y.min(), y.max(), len(set(y)))
    X, Y = np.meshgrid(xx, yy)
    Z = interpolate.griddata((x, y), z, (X, Y), method='cubic', rescale=0.3 )
    cbar=plt.cm.get_cmap('viridis').reversed()
    plt.pcolormesh(X,Y,Z, cmap=cbar)
    plt.plot(Gasfield['x'], Gasfield['y'], color='red')
    # plt.axis('equal')
    plt.colorbar()
    
    
```
